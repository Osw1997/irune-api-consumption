## Join conformed table with imports table

```python
df_hhi_production_egsehi = pd.read_csv("../data/hhi_production_exports_egsehi_conformed_table.csv")
final_imports_df = pd.read_csv("../data/imports_consolidated/imports_consolidated.csv")
```

```python
final_imports_df
```

```python
# Lets select "important" columns
# columns = [
#     "period", "reporterCode", "reporterISO", "reporterDesc", "flowDesc",
#     "partnerCode", "partnerISO", "partnerDesc", "classificationSearchCode",
#     "cmdCode", "cmdDesc", "qty", "netWgt", 
#     "qtyUnitAbbr", "cifvalue", "primaryValue"
# ]
columns = [
    "period", "reporterCode",
    "partnerCode", "partnerISO", "partnerDesc",
    "cmdCode", "qty", "netWgt", 
    "qtyUnitAbbr", "cifvalue", "primaryValue", 
    "netWgt_total_import", "cifvalue_total_import", "primaryValue_total_import"
]
final_imports_df = final_imports_df[columns]
final_imports_df.rename(
    columns={
        "period": "Year",
        "partnerCode": "partnerCode_import",
        "partnerISO": "partnerISO_import",
        "partnerDesc": "partnerDesc_import",
        "qty": "qty_import",
        "netWgt": "netWgt_import",
        "qtyUnitAbbr": "qtyUnitAbbr_import",
        "cifvalue": "cifvalue_import",
        "primaryValue": "primaryValue_import"
    }, inplace=True)

final_imports_df
```

```python
# Let's join both DFs using reporter code, product, year
all_data_df = pd.merge(df_hhi_production_egsehi, final_imports_df, how="outer", left_on=["reporterCode", "Year", "cmdCode"], right_on=["reporterCode", "Year", "cmdCode"])
all_data_df
```

## Calculate the Geopolitical risk

```python
all_data_df["Geopolitical Risk Production"] = all_data_df["HHI_production"] * (all_data_df["EGSEHI_6root"] * all_data_df["netWgt_import"] / (all_data_df["Domestic Production Value (P_AC)"] + all_data_df["netWgt_total_import"]))
all_data_df["Geopolitical Risk Exports"] = all_data_df["HHI_exports"] * (all_data_df["EGSEHI_6root"] * all_data_df["netWgt_import"] / (all_data_df["Domestic Production Value (P_AC)"] + all_data_df["netWgt_total_import"]))
all_data_df["Geopolitical Risk Only Political Stability Production"] = all_data_df["HHI_production"] * (all_data_df["value_governance_isolated"] * all_data_df["netWgt_import"] / (all_data_df["Domestic Production Value (P_AC)"] + all_data_df["netWgt_total_import"]))
all_data_df["Geopolitical Risk Only Political Stability Exports"] = all_data_df["HHI_exports"] * (all_data_df["value_governance_isolated"] * all_data_df["netWgt_import"] / (all_data_df["Domestic Production Value (P_AC)"] + all_data_df["netWgt_total_import"]))

all_data_df["Geopolitical Risk Production NO-ROOT"] = all_data_df["HHI_production"] * (all_data_df["EGSEHI"] * all_data_df["netWgt_import"] / (all_data_df["Domestic Production Value (P_AC)"] + all_data_df["netWgt_total_import"]))
all_data_df["Geopolitical Risk Exports NO-ROOT"] = all_data_df["HHI_exports"] * (all_data_df["EGSEHI"] * all_data_df["netWgt_import"] / (all_data_df["Domestic Production Value (P_AC)"] + all_data_df["netWgt_total_import"]))

all_data_df["Weighted Trave Average"] = all_data_df["netWgt_import"] / all_data_df["netWgt_total_import"]
all_data_df["~~ rojo ~~"] = all_data_df["Domestic Production Value (P_AC)"] / 1
```

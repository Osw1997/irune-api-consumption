```python
df_production_summarized = df_production.groupby(["Year", "Product"])[["Value"]].sum().reset_index()
df_production_summarized.rename(columns={"Value": "Total_value_YearProduct"}, inplace=True)
df_production_summarized
```

### Let's left-join production and total value of production by (Year, Product)

```python
df = pd.merge(df_production, df_production_summarized, how="left", left_on=["Year", "Product"], right_on=["Year", "Product"])
df
```

```python
df["Share in %"] = df["Value"] * 100 / df["Total_value_YearProduct"]
df["Share HHI Production"] = df["Share in %"] ** 2
df
```

### Sum the "Share HHI Production" by (Year, Product) and you get the HHI index :)

```python
df_hhi_production = df.groupby(["Year", "Product"])[["Share HHI Production"]].sum().reset_index()
df_hhi_production.rename(columns={"Share HHI Production": "HHI_production"}, inplace=True)
df_hhi_production = pd.merge(df, df_hhi_production, how="left", left_on=["Year", "Product"], right_on=["Year", "Product"])
df_hhi_production
# CONSIDERATIONS
# - Currently the HHI value is pasted to the row even if the production is null. Shall we replace by NANs if production is null?
```

```python
df_hhi_production.to_csv("../../data/hhi_data/df_hhi_production.csv", index=False)
```

## Now HHI for exports

```python
df_exports = pd.read_csv("../../data/exports/exports_consolidated_file.csv")
df_exports.rename(columns={"period": "Year"}, inplace=True)
df_exports
```

```python
df_exports_summarized = df_exports.groupby(["Year", "Product"])[["netWgt"]].sum().reset_index()
df_exports_summarized.rename(columns={"netWgt": "Total_netWgt_YearProduct"}, inplace=True)
df_exports_summarized
```

### Let's left-join production and total value of production by (Year, Product)

```python
df_e = pd.merge(df_exports, df_exports_summarized, how="left", left_on=["Year", "Product"], right_on=["Year", "Product"])
df_e
```

```python
df_e["Share in %"] = df_e["netWgt"] * 100 / df_e["Total_netWgt_YearProduct"]
df_e["Share HHI Exports"] = df_e["Share in %"] ** 2
df_e
```

### Sum the "Share HHI Exports" by (Year, Product) and you get the HHI index :)

```python
df_hhi_exports = df_e.groupby(["Year", "Product"])[["Share HHI Exports"]].sum().reset_index()
df_hhi_exports.rename(columns={"Share HHI Exports": "HHI_exports"}, inplace=True)
df_hhi_exports = pd.merge(df_e, df_hhi_exports, how="left", left_on=["Year", "Product"], right_on=["Year", "Product"])
df_hhi_exports

# SAME: Currently the HHI value is pasted to the row even if the production is null. Shall we replace by NANs if production is null?
```

```python
df_hhi_exports.to_csv("../../data/hhi_data/df_hhi_exports.csv", index=False)
```
